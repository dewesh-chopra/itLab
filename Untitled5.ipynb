**Plots in python**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.arange(10)
```

```python
x
```

```python
y = 4 * x + 5
```

```python
plt.plot(x, y)
plt.xlabel("X - Axis")
plt.ylabel("Y - Axis")
plt.show()
```

```python
x = np.arange(1, 11)
y = np.log(x)
plt.plot(x, y, "b-o")
plt.grid(True)
plt.show()
```

```python
x = np.arange(1, 101)
y = 3 * x * x + 5
plt.plot(x, y)
plt.grid(True)
plt.show()

x = np.arange(1, 101)
y = x * x * x - x * x - 1
plt.plot(x, y)
plt.grid(True)
plt.legend(["3x^2 + 5", "x^3 - x^2 - 1"])
plt.show()
```

```python
x = np.arange(1, 101)
y = np.cos(x)
plt.plot(x, y)
plt.grid(True)
plt.show()

x = np.arange(1, 101)
y = np.sin(x)
plt.plot(x, y)
plt.grid(True)
plt.legend(["3x^2 + 5", "x^3 - x^2 - 1"])
plt.title("Plot of y1 = 3x^2 + 5 and y2 = x^3 - x^2 - 1")
plt.show()
```

```python
x = np.arange(1, 101)
y = np.tan(x)
plt.plot(x, y)
plt.grid(True)
plt.gcf().set_size_inches(10, 10)
plt.gca().set_xticks(np.linspace(1, 101, 20))
# plt.gca().set_yticks(np.linspace(1, 101, 20))
plt.show()
```

```python
x = np.arange(1, 101)
y = 3 * (1 + np.cos(x))
plt.plot(x, y)
plt.show()
```

```python
#Produce theta variable
theta = np.arange(0, 2*np.pi, .01)[1:]

#Input polar equation of 2sinθ + 2cosθ 
r = 2 * np.cos(theta) + 2 * np.sin(theta) 
# Adding "()" around eq doesn't change anything

#Make plt figure
fig = plt.figure()

#Make sub-plot with attribute "polar"
ax = fig.add_subplot(polar=True)

#Plot function
ax.plot(theta, r)

#Show plot
plt.show()
```

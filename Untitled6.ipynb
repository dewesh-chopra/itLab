```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

x = np.arange(1, 4)
y1 = x
y2 = x * x
y3 = x * x * x
y4 = np.log(x)

plt.plot(x, y1, "c", x , y2, "m", x, y3, "y", x, y4, "k")
plt.grid(True)
plt.legend(["y = x", "y = x^2", "y = x^3", "y = log(x)"])
plt.title("Some plots")
plt.xlabel("x - axis")
plt.ylabel("y - axis")
plt.gcf().set_size_inches(10, 10)
plt.gca().set_yticks(np.linspace(1, 30, 10))
plt.show()
```

```python
x = np.linspace(-np.pi, np.pi, 100)
y = np.sin(np.exp(x * x))
plt.xlabel("x - axis")
plt.ylabel("y - axis")
plt.title("Plot 1")
plt.grid(True)
plt.plot(x, y, "m")
plt.show()
```

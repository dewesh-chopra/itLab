```python
x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x**2)

fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title('Simple plot')
fig.suptitle('Examples')
plt.show()
```

```python
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x**2)+np.cos(x**2)

fig, axs = plt.subplots(2, 2, subplot_kw = dict(polar = True))
axs[0, 0].plot(x, y)
axs[0, 1].scatter(x, y)
axs[1, 0].scatter(x, y)
axs[1, 1].plot(x, y)

fig.suptitle('Example')
plt.show()
```

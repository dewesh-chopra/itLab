```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
iris = np.genfromtxt("C:\\Users\\BIT.L1-057\\Desktop\\iris.numeric", delimiter = ",")
```

```python
iris = pd.DataFrame(iris)
iris
```

```python
iris.columns = ["SL", "SW", "PL", "PW", "Species"]
```

```python
colors = ["red", "blue", "yellow"]
for i in range(len(colors)):
    plt.scatter(iris[iris["Species"] == i+1]["SL"], iris[iris["Species"] == i+1]["PL"])
plt.show()
```

```python
plt.boxplot(iris["SL"])
plt.show()
```

```python
fig, ax = plt.subplots(nrows = 1, ncols = 2)
ax[0].hist(iris["SL"])
ax[1].hist(iris["PL"])
plt.show()
```

```python
fig, ax = plt.subplots(nrows = 1, ncols = 2, sharex = True, sharey = True)
ax[0].hist(iris["SL"])
ax[1].hist(iris["PL"])
plt.show()
```

```python
plt.scatter(housing["latitude"], housing["longitude"])
```

```python
seasons = ["summer", "winter", "autumn", "spring"]
for i in seasons:
    plt.scatter(housing["latitude"], housing["longitude"], s = housing["population"]/25, c = housing["median_house_value"], cmap = i, alpha = 0.7)
    plt.xlabel("longitude")
    plt.ylabel("latitude")
    plt.colorbar()
    plt.show()
```
